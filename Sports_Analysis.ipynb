
Calculate the win and loss ratio for each team in the dataset "EPL_Set.csv".

```python
all_teams = pd.concat([df['HomeTeamName'], df['AwayTeamName']]).unique()
team_stats = {team: {'Wins': 0, 'Losses': 0} for team in all_teams}
```

loop through each row of the DataFrame and access the required columns.



```python
for index, row in df.iterrows():
    result = row['Result']
    home_team = row['HomeTeamName']
    away_team = row['AwayTeamName']
```

```python
for index, row in df.iterrows():
    result = row['Result']
    home_team = row['HomeTeamName']
    away_team = row['AwayTeamName']

    if result == 'H':
        team_stats[home_team]['Wins'] += 1
        team_stats[away_team]['Losses'] += 1
    elif result == 'A':
        team_stats[away_team]['Wins'] += 1
        team_stats[home_team]['Losses'] += 1
```

Calculate the win and loss ratios for each team and store them in the team_stats dictionary.



```python
for team, stats in team_stats.items():
    total_games = stats['Wins'] + stats['Losses']
    if total_games > 0:
        team_stats[team]['WinRatio'] = stats['Wins'] / total_games
        team_stats[team]['LossRatio'] = stats['Losses'] / total_games
    else:
        team_stats[team]['WinRatio'] = 0
        team_stats[team]['LossRatio'] = 0

display(pd.DataFrame.from_dict(team_stats, orient='index'))
```


Convert the team_stats dictionary to a pandas DataFrame and display it.



```python
team_stats_df = pd.DataFrame.from_dict(team_stats, orient='index')
display(team_stats_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
team_season_stats = {}
```

```python
for index, row in df.iterrows():
    season = row['Season']
    home_team = row['HomeTeamName']
    away_team = row['AwayTeamName']
    home_score = row['HomeTeamScore']
    away_score = row['AwayTeamScore']

    if season not in team_season_stats:
        team_season_stats[season] = {}

    if home_team not in team_season_stats[season]:
        team_season_stats[season][home_team] = {'Wins': 0, 'Losses': 0, 'GoalsScored': 0, 'GoalsConceded': 0}
    if away_team not in team_season_stats[season]:
        team_season_stats[season][away_team] = {'Wins': 0, 'Losses': 0, 'GoalsScored': 0, 'GoalsConceded': 0}

    if row['Result'] == 'H':
        team_season_stats[season][home_team]['Wins'] += 1
        team_season_stats[season][away_team]['Losses'] += 1
    elif row['Result'] == 'A':
        team_season_stats[season][away_team]['Wins'] += 1
        team_season_stats[season][home_team]['Losses'] += 1

    team_season_stats[season][home_team]['GoalsScored'] += home_score
    team_season_stats[season][home_team]['GoalsConceded'] += away_score
    team_season_stats[season][away_team]['GoalsScored'] += away_score
    team_season_stats[season][away_team]['GoalsConceded'] += home_score
```

**Reasoning**:
Calculate total games played, win ratio, loss ratio, average goals scored, and average goals conceded for each team in each season and convert the aggregated data into a pandas DataFrame.



```python
aggregated_data = []
for season, teams_data in team_season_stats.items():
    for team, stats in teams_data.items():
        total_games = stats['Wins'] + stats['Losses']
        win_ratio = stats['Wins'] / total_games if total_games > 0 else 0
        loss_ratio = stats['Losses'] / total_games if total_games > 0 else 0
        avg_goals_scored = stats['GoalsScored'] / total_games if total_games > 0 else 0
        avg_goals_conceded = stats['GoalsConceded'] / total_games if total_games > 0 else 0

        aggregated_data.append({
            'Season': season,
            'Team': team,
            'Wins': stats['Wins'],
            'Losses': stats['Losses'],
            'GoalsScored': stats['GoalsScored'],
            'GoalsConceded': stats['GoalsConceded'],
            'GamesPlayed': total_games,
            'WinRatio': win_ratio,
            'LossRatio': loss_ratio,
            'AvgGoalsScored': avg_goals_scored,
            'AvgGoalsConceded': avg_goals_conceded
        })

team_season_stats_df = pd.DataFrame(aggregated_data)
display(team_season_stats_df.head())
```

Generating line graph over team perfomances

```python
selected_teams = ['Man United', 'Liverpool', 'Arsenal', 'Chelsea']

plt.figure(figsize=(12, 8))

for team in selected_teams:
    team_data = team_season_stats_df[team_season_stats_df['Team'] == team]
    plt.plot(team_data['Season'], team_data['WinRatio'], marker='o', linestyle='-', label=team)

plt.title('Win Ratio Over Seasons for Selected Teams')
plt.xlabel('Season')
plt.ylabel('Win Ratio')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
